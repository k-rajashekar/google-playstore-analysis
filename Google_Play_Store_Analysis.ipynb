## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
playstore_data.drop_duplicates(inplace=True)
playstore_data.duplicated().sum()
userreview_data.drop_duplicates(inplace=True)
userreview_data.duplicated().sum()
````

````python
playstore_data.replace('null', np.nan, inplace=True)
userreview_data.replace('null', np.nan, inplace=True)
playstore_data.isnull().sum()
userreview_data.isnull().sum()
playstore_data.head()
````

````python
playstore_data.loc[:, 'Rating'].fillna(playstore_data['Rating'].mean(), inplace=True)
playstore_data.loc[:, 'Type'].fillna(playstore_data['Type'].mode()[0], inplace=True)
playstore_data.loc[:, 'Content Rating'].fillna(playstore_data['Content Rating'].mode()[0], inplace=True)
playstore_data.loc[:, 'Current Ver'].fillna(playstore_data['Current Ver'].mode()[0], inplace=True)
playstore_data.loc[:, 'Android Ver'].fillna(playstore_data['Android Ver'].mode()[0], inplace=True)
playstore_data.isnull().sum()
````

````python
userreview_data.isnull().sum()
userreview_data.head(1)
userreview_data.dropna(subset=['Translated_Review', 'Sentiment', 'Sentiment_Polarity', 'Sentiment_Subjectivity'], inplace=True)
userreview_data.isnull().sum()
````

````python
playstore_data['Installs'] = playstore_data['Installs'].str.replace('[+,]', '', regex=True).str.strip()

# Check for any non-numeric values that might still exist
non_numeric_installs = playstore_data[~playstore_data['Installs'].str.isnumeric()]
print(non_numeric_installs['Installs'].unique())
non_numeric_price = playstore_data[~playstore_data['Price'].str.isnumeric()]
print(non_numeric_price['Price'].unique())
````

````python
# Cleaning 'Installs' column
# Removing ',' and '+' symbols, then converting to float for numeric analysis
# Replace 'Free' with 0, as it indicates no price or a free app
playstore_data['Installs'] = playstore_data['Installs'].replace('Free', 0)

# Remove ',' and '+' using regex and strip spaces
playstore_data['Installs'] = playstore_data['Installs'].str.replace('[+,]', '', regex=True).str.strip()

# Convert to float safely
playstore_data['Installs'] = pd.to_numeric(playstore_data['Installs'], errors='coerce')

# Cleaning 'Price' column
# Removing '$' symbol and converting to float
playstore_data['Price'] = playstore_data['Price'].str.replace('Everyone', '0')
playstore_data['Price'] = playstore_data['Price'].str.replace('$', '', regex=False)
playstore_data['Price'] = playstore_data['Price'].astype(float)

# Cleaning 'Size' column
# Converting 'Size' to consistent MB values:
# - 'k' â†’ kilobytes to MB (divided by 1024)
# - 'M' â†’ megabytes as is
# - 'G' â†’ gigabytes to MB (multiplied by 1024)
# - 'Varies with device' â†’ set as NaN

def convert_size(size_str):
    if 'M' in size_str:
        return float(size_str.replace('M', ''))
    elif 'k' in size_str:
        return float(size_str.replace('k', '')) / 1024
    elif 'G' in size_str:
        return float(size_str.replace('G', '')) * 1024
    else:
        return np.nan  # for 'Varies with device' or unexpected entries

# Replace 'Varies with device' with NaN first
playstore_data['Size'] = playstore_data['Size'].replace('Varies with device', np.nan)

# Drop NaNs (optional: or keep and impute later), and apply size conversion
playstore_data['Size'] = playstore_data['Size'].dropna().apply(convert_size)

# Check the results
playstore_data[['Installs', 'Price', 'Size']].info()
playstore_data[['Installs', 'Price', 'Size']].describe()
````

### Write all manipulations have you done and insights you found?

To prepare the dataset for accurate visualizations and analysis, the following data wrangling steps were performed:

Duplicate records were removed from both datasets to eliminate redundancies and ensure data accuracy.

All occurrences of the string 'null' were replaced with proper NaN values for appropriate handling using pandas.

Missing values in the Play Store dataset were imputed:

Rating was filled using the column's mean value.

Categorical fields like Type, Content Rating, Current Ver, and Android Ver were filled using their mode values to preserve consistency.

In the User Reviews dataset, rows lacking Translated_Review, Sentiment, Sentiment_Polarity, or Sentiment_Subjectivity were dropped â€” as these are core features for sentiment analysis.

Numeric cleanup of important fields:

Installs: Cleaned by removing commas and plus signs, and converted to numerical format for aggregation and visualization.

Price: Dollar signs were removed, and values were converted to float.

Size: Handled inconsistencies by converting values like k, M, and G to megabytes (MB), and treating "Varies with device" as missing.
